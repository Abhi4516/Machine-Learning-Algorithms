```python
dt_model2=[]
for ccp in ccp_alphas:
    dt_m=DecisionTreeClassifier(ccp_alpha=ccp)
    dt_m=dt_m.fit(x_train,y_train)
    dt_model2.append(dt_m)
```

```python
dt_model2
```

```python
plt.show()
```

```python
train_score = [i.score(x_train,y_train) for i in dt_model2]
```

```python
test_score = [i.score(x_test ,y_test) for i in dt_model2]
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_score, marker='o', label="train",
        drawstyle="steps-post")
ax.plot(ccp_alphas, test_score, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```
